```python
import shapes
import matplotlib.pyplot as plt
import numpy as np
import pulse2percept
```

```python
fig, axs = plt.subplots(ncols=1, figsize=(15, 15))

s3 = shapes.subject_params['51-009']
implant, model = shapes.model_from_params(s3, biphasic=False)
implant.plot(annotate=True,ax=axs)
model.plot(ax=axs)

axs.set(ylim=(-5000,5000),xlim=(-6000,6000))
```
